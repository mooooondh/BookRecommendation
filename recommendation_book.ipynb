```python
import pandas as pd
import numpy as np
```

## load data

```python
df_user_data= pd.read_csv("./data/user_data.csv")
```

```python
df_user_data.head()
```

```python
df_user_data= df_user_data[["ID", "Name", "Rating"]]
df_user_data.head()
```

## translate user-book matrix

```python
rating= df_user_data[["ID", "Name", "Rating"]]
rating_matrix= rating.pivot_table("Rating", index= "ID", columns= "Name").fillna(0)
rating_matrix.head()
```

```python
rating_matrix_tr= rating_matrix.transpose()
rating_matrix_tr.head()
```

## make book similarity matrix

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim= cosine_similarity(rating_matrix_tr, rating_matrix_tr)

df_sim= pd.DataFrame(data= sim, index= rating_matrix.columns, columns= rating_matrix.columns)
df_sim.head()
```

## identification of highly relevant books

```python
df_sim["\"A Problem from Hell\": America and the Age of Genocide"].sort_values(ascending= False)[:11]
```

## make recommendation matrix

```python
def predict_rating(rating, item_sim):
    rating_pred= rating.dot(item_sim)/np.array([np.abs(item_sim).sum(axis= 1)])
    return rating_pred
```

```python
rating_pred= predict_rating(rating_matrix.values, df_sim.values)
rating_pred_matrix= pd.DataFrame(data= rating_pred, index= rating_matrix.index, columns= rating_matrix.columns)
rating_pred_matrix.head()
```

## check MSE

```python
from sklearn.metrics import mean_squared_error
```

```python
def mse(pred, actual):
    pred= pred[actual.nonzero()].flatten()
    actual= actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)
```

```python
print("mse: ", mse(rating_pred, rating_matrix.values))
```

## reduce MSE

```python
def predict_rating_top_n(rating, item_sim, n= 20):
    pred= np.zeros(rating.shape)
    
    for col in range(rating.shape[1]):
        top_n_item= [np.argsort(item_sim[:, col])[:-n-1:-1]]
        for row in range(rating.shape[0]):
            pred[row, col]= item_sim[col, :][top_n_item].dot(rating[row, :][top_n_item].T)
            pred[row, col]/= np.sum(np.abs(item_sim[col, :][top_n_item]))
            
    return pred
```

```python
rating_pred_2= predict_rating_top_n(rating_matrix.values, df_sim.values, n= 20)
```

```python
print("mse: ", mse(rating_pred_2, rating_matrix.values))
```

```python
rating_pred_matrix= pd.DataFrame(data= rating_pred_2, index= rating_matrix.index, columns= rating_matrix.columns)
rating_pred_matrix.head()
```

## check all books tha user#1 doesn't read

```python
def get_user_unread(rating, use_id):
    user_rating= rating.loc[use_id, :]
    already_read= user_rating[user_rating> 0].index.tolist()
    book_list= rating.columns.tolist()
    unread_list= [book for book in book_list if book not in already_read]
    return unread_list
```

## recommendation 10books for user#1

```python
def recommendation(pred_df, use_id, unread_list, top_n= 10):
    re_book= pred_df.loc[use_id, unread_list].sort_values(ascending= False)[:top_n]
    return re_book
```

```python
unread_list= get_user_unread(rating_matrix, 9)
```

```python
re_book= recommendation(rating_pred_matrix, 9, unread_list, top_n= 10)
```
